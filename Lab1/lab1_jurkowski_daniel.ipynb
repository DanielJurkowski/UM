```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn.datasets import load_iris
irisRaw = load_iris()
```

#### Zadanie 1

```python
iris = pd.DataFrame(data= np.c_[irisRaw['data'], irisRaw['target']],
                     columns= irisRaw['feature_names'] + ['target'])
```

#### Zadanie 11

```python
print(iris['target'].unique())
```

```python
columns = list(iris.columns.values)[:-1]

for column in columns:
    plt.hist(iris[iris['target'] == 0][column])
    plt.hist(iris[iris['target'] == 1][column])
    plt.hist(iris[iris['target'] == 2][column])
    plt.grid()
    plt.title(column)
    plt.xlabel('value')
    plt.ylabel('count')
    plt.legend(['target = 0.0', 'target = 1.0', 'target = 2.0'])
    plt.show()
```
